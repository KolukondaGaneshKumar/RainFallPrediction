```python
def outlier(fechur):
    half=df[fechur].quantile(0.75)-df[fechur].quantile(0.25)
    left=df[fechur].quantile(0.25)-half*1.5
    right=df[fechur].quantile(0.75)+half*1.5
    df.loc[df[fechur]>=right, fechur]=right
    df.loc[df[fechur]<=left, fechur]=left
```

```python
for x in con_feat:
    outlier(x)
```
